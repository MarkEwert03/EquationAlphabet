```python
def listprint(mylist: list, k: int = 1, newline: bool = True):
    for i in range(0, len(mylist), k):
        print(mylist[i : i + k])
    if newline:
        print()


print("GENERATING ALL EQUATIONS")
print("n = 1:")
listprint(generate_all_equations(1), 2)
print("n = 2")
listprint(generate_all_equations(2), 2)
print("n = 3")
listprint(generate_all_equations(3), 6)
```
